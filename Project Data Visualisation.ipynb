```python
 df1 = df1.rename(columns={'FAVC': 'frequent_cons_high_caloric_food', 'FCVC': 'frequency_cons_veg', 
                           'NCP': 'nb_main_meal', 'CAEC': 'cons_between_meal', 
                           'SCC': 'Suivi_calories', 'FAF': 'frequency_phys_act', 
                           'TUE': 'time_tech', 'CALC': 'cons_alcohol'})
```

```python
df1
```

```python
df1['family_history_with_overweight'] = df1['family_history_with_overweight'].replace(['yes','no'],[1,0])
df1['frequent_cons_high_caloric_food'] = df1['frequent_cons_high_caloric_food'].replace(['yes','no'],[1,0])
df1['cons_between_meal'] = df1['cons_between_meal'].replace(['no','Sometimes','Frequently','Always'],[0,1,2,3])
df1['SMOKE'] = df1['SMOKE'].replace(['yes','no'],[1,0])
df1['Suivi_calories'] = df1['Suivi_calories'].replace(['yes','no'],[1,0])
df1['MTRANS'] = df1['MTRANS'].map({'Public_Transportation':0,
                             'Walking':1,'Automobile':2,
                             'Motorbike':3,'Bike':4},na_action=None)
df1['cons_alcohol'] = df1['cons_alcohol'].map({'no':0,
                             'Sometimes':1,'Frequently':2,
                             'Always':3},na_action=None)
```

### Data visualisation

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
columns = ['Gender', 'family_history_with_overweight']
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
for col, subplot in zip(columns, ax.flatten()):
    sns.countplot(df1[col], ax=subplot)
    sns.countplot(df1[col],ax=subplot)  
    subplot.yaxis.label.set_text("Number of Records")
        
# show figure & plots
plt.tight_layout(pad=5, w_pad=0.0, h_pad=1)
plt.show()
```

```python
dfMale=df1[df1['Gender']=='Male']
len(dfMale)
```

```python
dfFemale=df1[df1['Gender']=='Female']
len(dfFemale)
```

```python
dftest=df1.groupby('Gender').count()
```

```python
df1.groupby('Gender').size().plot(kind='pie', autopct="%.1f%%", textprops={'fontsize': 20},
                                 colors=['tomato', 'royalblue'])

plt.ylabel('Per gender', size=22)
plt.tight_layout()
plt.show()
```

```python
df1.groupby('SMOKE').size().plot(kind='pie', autopct="%.1f%%", textprops={'fontsize': 12},
                                 colors=['violet', 'lime'])

plt.ylabel('Smoke', size=18)
plt.tight_layout()
plt.show()
```

```python
df1.groupby('NObeyesdad').size().plot(kind='pie', autopct="%.1f%%", textprops={'fontsize': 12})

plt.ylabel('Obesity Types', size=18)
plt.tight_layout()
plt.show()
```
